## Step 2: Handle Missing Values

```python
# Check missing values per column
missing = df.isnull().sum()
missing[missing > 0]
```

```python
print(f"Total missing values: {df.isnull().sum().sum()}")
```

```python
# Fill 'discount' (numeric) missing values with 0 (means no discount was given)
df['discount'] = df['discount'].fillna(0)
```

```python
# Verify no missing values left
print(f"Missing values now: {df.isnull().sum().sum()}")
```

## Step 3: Basic Operations (Filter Rows, Select Columns)

```python
# Filter: keep only products with rating >= 1.0
before = len(df)
df = df[df['rating'] >= 1.0]
print(f"Filtered rating >= 1.0: {before} -> {len(df)} rows")
```

```python
# Filter: keep only products with price > 0
before = len(df)
df = df[df['initial_price'] > 0]
print(f"Filtered initial_price > 0: {before} -> {len(df)} rows")
```

```python
# Select only the important columns
selected_cols = ['product_id', 'title', 'product_description', 'rating',
                 'ratings_count', 'initial_price', 'discount', 'currency', 'category']
df = df[selected_cols]
print(f"Selected {len(selected_cols)} columns")
df.head()
```

## Step 4: Remove Duplicates

```python
print(f"Duplicate rows: {df.duplicated().sum()}")
```

### That is , no need to remove duplicates

## Step 5: Create Derived Columns

```python
# Derived Column 1: discount_amount (actual rupee savings)
df['discount_amount'] = df['initial_price'] * (df['discount'] / 100)

# Derived Column 2: value_score (composite of rating & popularity)
df['value_score'] = df['rating'] * df['ratings_count']

print(f"Added 2 derived columns. Total columns now: {len(df.columns)}")
print(f"Columns: {list(df.columns)}")
df[['title', 'initial_price', 'discount', 'discount_amount', 'rating', 'ratings_count', 'value_score']].head(10)
```

## Step 6: Save Cleaned Dataset

```python
df.to_csv("Cleaned_dataset.csv", index=False)
print(f"Saved 'Cleaned_dataset.csv'")
print(f"Final shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
```
